```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from apyori import apriori
```

```python
#Importamos los datos a usar
path = 'tienda.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.drop(['Item(s)'], axis=1,inplace=True) ##Eliminamos la columna inservible
```

```python
df
```

```python
df.shape  #vemos la forma de nuestros datos, tenemos 9835 registros y 32 columnas
```

```python
records=[]
for i in range(0,9835):
    records.append([str(df.values[i,j]) for j in range(0,32)]) 
```

```python
records #Para usar la libreria necesitamos poner los datos en forma de lista
```

```python
association_rules = apriori(records, min_support=.003050,min_confidence=0.2,min_lift=3,min_length=2)   #30/9835 = 0.003050
```

```python
association_results=list(association_rules) 
```

```python
for element in association_results:
    print (element)
```

```python
association_results[0]
#Podemos ver que el renglon  0 los productos hambuger meat y Instant foodproducts
#usualmente se compran juntos

#el nivel de confianza es de .20895, esto nos dice que el 20%  de las compras que contiene
#el producto A tendran en su lista el producto B

#El lift de 3.79 nos dice que el producto A tiene mas probabilidades de ser comprado 
#por aquellos clientes qye tambien compran el producto B


#De esta manera podriamos hacer el analisis de cada una de las reglas que se encontraron c:
```

```python
#El siguiente codigo nos permite ver las reglas encontradas de una mejor manera!

results = []
for item in association_results:

  
    pair =item [0]
    items= [x for x in pair]
    value0 = str(items[0])
    value1 = str(items[1])
    value3 = str(item[2][0][2])[:7]
    value4 = str(item[2][0][3])[:7]

    rows = (value0, value1, value3, value4)

    results.append(rows)

labels = ['Producto A', 'Producto B', 'Confidence', 'Lift'] 
movie_suggestion = pd.DataFrame.from_records (results, columns = labels)

print(movie_suggestion)
```
